```python
import matplotlib.pyplot as plt
```

```python
# Taken from tensorboard logs CNN14_BASE confusionmatrices, class Accordion

# Threshold = 0
TP_0 = 7
TN_0 = 0
FP_0 = 987
FN_0 = 0

# Threshold = 0.1
TP_10 = 7
TN_10 = 315
FP_10 = 627
FN_10 = 0

# Threshold = 0.25
TP_25 = 7
TN_25 = 746
FP_25 = 240
FN_25 = 1

# Threshold = 0.5
TP_50 = 6
TN_50 = 987
FP_50 = 0
FN_50 = 1

# Threshold = 0.75
TP_75 = 2
TN_75 = 987
FP_75 = 0
FN_75 = 5

# Threshold = 0.9
TP_90 = 1
TN_90 = 987
FP_90 = 0
FN_90 = 6

# Threshold = 1
TP_100 = 0
TN_100 = 987
FP_100 = 0
FN_100 = 7
```

```python
# Calculate TPR and FPR

TPR_0 = TP_0 / (TP_0 + FN_0)
FPR_0 = FP_0 / (FP_0 + TN_0)

TPR_10 = TP_10 / (TP_10 + FN_10)
FPR_10 = FP_10 / (FP_10 + TN_10)

TPR_25 = TP_25 / (TP_25 + FN_25)
FPR_25 = FP_25 / (FP_25 + TN_25)

TPR_50 = TP_50 / (TP_50 + FN_50)
FPR_50 = FP_50 / (FP_50 + TN_50)

TPR_75 = TP_75 / (TP_75 + FN_75)
FPR_75 = FP_75 / (FP_75 + TN_75)

TPR_90 = TP_90 / (TP_90 + FN_90)
FPR_90 = FP_90 / (FP_90 + TN_90)

TPR_100 = TP_100 / (TP_100 + FN_100)
FPR_100 = FP_100 / (FP_100 + TN_100)


TPR = [TPR_0, TPR_10, TPR_25, TPR_50, TPR_75, TPR_90, TPR_100]
FPR = [FPR_0, FPR_10, FPR_25, FPR_50, FPR_75, FPR_90, FPR_100]
```

```python
# Plot the ROC curve
plt.figure()
plt.plot(FPR, TPR, marker='o', linestyle='-', color='b', label='Accordion')
plt.plot([0, 1], [0, 1], linestyle='--', color='r')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
