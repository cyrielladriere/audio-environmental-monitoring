```python
import os
import wave
import contextlib
import matplotlib.pyplot as plt

# Path to the folder containing .wav files
folder_path = "../data/audio/train_curated/"
```

```python
# Function to get the duration of a .wav file
def get_wav_duration(wav_path):
    with contextlib.closing(wave.open(wav_path, 'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration
```

```python
# List to store the durations
durations = []

# Iterate over all .wav files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.wav'):
        file_path = os.path.join(folder_path, filename)
        duration = get_wav_duration(file_path)
        durations.append(duration)
```

```python
# Plot the distribution of durations
plt.hist(durations, bins=range(0, 31), edgecolor='black')
plt.xlabel('Duration (seconds)')
plt.ylabel('Number of audio clips')
plt.title('Distribution of DCASE 2019 audio clip durations')
plt.show()
```
